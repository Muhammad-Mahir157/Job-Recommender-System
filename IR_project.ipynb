**Jobs Recommender System Based on Popularity**

```python
#merging preferrences with Jobs based on Title
ratings_with_name = ratings.merge(jobs,on='id')
```

```python
ratings_with_name.head(10)
```

```python
# Getting the count for each job-Title using group-by
num_rating_df = ratings_with_name.groupby('Title_x').count()['benefits.numRatings'].reset_index()
num_rating_df
```

```python
# Getting the Mean/Avg for each Job-Title using group-by
avg_rating_df = ratings_with_name.groupby('Title_x').mean()['benefits.numRatings'].reset_index()
avg_rating_df
```

```python
#renaming the rating columns
num_rating_df.rename(columns={'benefits.numRatings':'num_ratings'},inplace=True)
avg_rating_df.rename(columns={'benefits.numRatings':'avg_rating'},inplace=True)
```

```python
# Merging the count-df with the avg-df based on the Job-Title
popular_df = num_rating_df.merge(avg_rating_df,on='Title_x')
popular_df
```

```python
# Fetching the jobs having more than 30 ratings and sorting them
popular_df = popular_df[popular_df['num_ratings']>=30].sort_values('avg_rating',ascending=False)
```

```python
print(len(popular_df))
```

```python
jobs.rename(columns={'gaTrackerData.jobTitle':'Title_x'},inplace=True)
```

**Performing Collaborative Filtering**

```python
#Considering only those users who have rated more than 10 jobs
temp_X = ratings_with_name.groupby('id').count()['benefits.numRatings'] > 10
potential_employees = temp_X[temp_X].index
```

```python
print(potential_employees)
```

```python
#Fetching the filtered users uing the id's of the users
filtered_rating = ratings_with_name[ratings_with_name['id'].isin(potential_employees)]
```

```python
print(filtered_rating)
```

```python
#Considering only those books that have been rated at least 30 times
temp_Y = filtered_rating.groupby('Title_x').count()['benefits.numRatings']>=30
famous_designations = temp_Y[temp_Y].index
```

```python
print(famous_designations)
```

```python
#Fetching the filtered books uing the Book title
final_ratings = filtered_rating[filtered_rating['Title_x'].isin(famous_designations)]
```

```python
# Creating the table from the filtered data
pt = final_ratings.pivot_table(index='Title_x',columns='id',values='benefits.numRatings')
```

```python
#filling the null values with 0
pt.fillna(0,inplace=True)
```

```python
pt.head(20)
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(designation):
    # index fetch using the designation
    index = np.where(pt.index==designation)[0][0]
    # Get the top 5 jobs with max similarity scores
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = jobs[jobs['Title_x'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Title_x')['Title_x'].values))
        item.extend(list(temp_df.drop_duplicates('Title_x')['gaTrackerData.empName'].values))
        item.extend(list(temp_df.drop_duplicates('Title_x')['gaTrackerData.industry'].values))
        item.extend(list(temp_df.drop_duplicates('Title_x')['overview.website'].values))
        #'Title','gaTrackerData.empName','gaTrackerData.industry','num_ratings','avg_rating','overview.website']]
        data.append(item)
    
    return data
```

```python
recommend('Business Analyst')
```
